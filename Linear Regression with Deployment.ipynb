```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

## Load Boston House Pricing Dataset

```python
from sklearn.datasets import fetch_openml
boston = fetch_openml(name='boston', version=1)
```

```python
boston.keys()
```

```python
boston.DESCR
```

## Preparing the Dataset

```python
df = pd.DataFrame(boston.data)
```

```python
df
```

```python
df['Price'] = boston.target
df.head()
```

```python
df.info()
```

```python
df['CHAS'] = df['CHAS'].astype(float)
df['RAD'] = df['RAD'].astype(float)
df.info()
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
# Independent and Dependent Feature

X = df.drop('Price', axis=1)
y = df['Price']
```

```python
X.head()
```

```python
y.head()
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
# Standardize the data

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
X_test_scaled
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

```python
# Print Coefficient and Intercept
print(regression.coef_)
print(regression.intercept_)
```

```python
# Parameters the model has been trained
print(regression.get_params())
```

```python
# Prediction with test data
regression_pred = regression.predict(X_test)
regression_pred
```

```python
# Plot scatter plot for prediction
plt.scatter(regression_pred, y_test)
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Scatter Plot of Predicted vs Actual Prices')
plt.show()
```

```python
# Plotting residuals
residuals = y_test - regression_pred
plt.scatter(regression_pred, residuals)
plt.axhline(y=0, color='r', linestyle='--')
plt.xlabel('Predicted Price')
plt.ylabel('Residuals')
# Uniform Distribution
```

```python
sns.displot(residuals, kind="kde")
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
print('MSE:', mean_squared_error(y_test, regression_pred))
print('MAE:', mean_absolute_error(y_test, regression_pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test, regression_pred)))
print('R2:', r2_score(y_test, regression_pred))
print('Adjusted R:', 1 - (1 - r2_score(y_test, regression_pred)) * (len(y_test) - 1) / (len(y_test) - X_test.shape[1] - 1))
```
